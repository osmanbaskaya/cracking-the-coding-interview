```python
### Solution for Question 1.2 ###

def map2array(text):
    start_char_val = ord('A')
    size = ord('z') - start_char_val
    array = [0] * size
    for c in text:
        val = ord(c) - start_char_val
        assert 0 <= val < size, "Character out of range"
        array[val] += 1

    return array
        
def compare_arrays_equal(arr1, arr2):
    for i in range(len(arr1)):
        if arr1[i] != arr2[i]:
            return False
    return True


assert compare_arrays_equal(map2array('dog'), map2array('god'))
assert compare_arrays_equal(map2array('dogg'), map2array('goda')) == False
assert compare_arrays_equal(map2array(''), map2array(''))
```

```python
### Solution for Question 1.3 ###

def is_space(c):
    return c == ' '


def replace_space(text, replacement='%20'):
    length_of_rep = len(replacement)
    for i in range(len(text)):
        if is_space(text[i]):
            slide(text, i, length_of_rep - 1)
            for j in range(i, i + length_of_rep):
                text[j] = replacement[j-i]
            i = j
    return text

def slide(text, i, length_of_rep):
    trailing_skipped = False
    for j in range(len(text)-1, i, -1):
        if is_space(text[j]) and not trailing_skipped:
            continue
        else:
            trailing_skipped = True
            text[j+length_of_rep] = text[j]


assert "".join(replace_space(list("Osman  Ba skay a        "))) == 'Osman%20%20Ba%20skay%20a'

### Another Solution for 1.3 ###

def replace_space2(text, true_length, replacement='%20'):
    num_space = 0
    for i in range(true_length):
        if text[i] == ' ':
            num_space += 1
    
    length_of_rep = len(replacement)
    index = true_length + 2 * num_space
    
    for i in range(true_length - 1, -1, -1):
        if text[i] == ' ':
            for j in range(length_of_rep - 1, -1, -1):
                index -= 1
                text[index] = replacement[j]
        else:
            index -= 1
            text[index] = text[i]
            
    return text  

assert "".join(replace_space2(list("Mr John Smith    "), 13)) == 'Mr%20John%20Smith'
```

```python
### Solution for Question 1.4 ###
from collections import Counter

def replace_space(text):
    # return text.replace(' ', '')
    new = ''
    for c in text:
        if c != ' ':
            new = "{}{}".format(new, c)
    return new

def check_perm_palindrom(text):
    text = replace_space(text)
    text = text.lower()
    freqs = Counter(text)
    num_of_odds = 0
    for v in freqs.values():
        if v % 2 == 1:
            num_of_odds += 1
            
    if len(text) % 2 == 0:
        return num_of_odds == 0
    else:
        return num_of_odds <= 1
        

print(check_perm_palindrom("taco  Caat"))
```

```python
### Solution for Question 1.5 ###


def __check_one_away_equal(str1, str2, max_mistake=1):
    num_of_unmatch = 0
    for c1, c2 in zip(str1, str2):
        if c1 != c2:
            num_of_unmatch += 1
        if num_of_unmatch > 1:
            break

    return num_of_unmatch <= max_mistake


def __check_one_away_small(str1, str2, max_mistake=1):
    if len(str2) - len(str1) > max_mistake:
        return False

    i = 0
    num_of_unmatch = 0
    for j in range(len(str1)):
        for k in range(max_mistake + 1):
            if str1[j] == str2[i]:
                break
            else:
                num_of_unmatch += 1
                i += 1
        if num_of_unmatch > max_mistake:
            break
        i += 1

    return num_of_unmatch <= max_mistake
        

def check_one_away(str1, str2, max_mistake=1):

    ans = None

    if str1 == str2:
        ans = True
    else:
        if len(str1) == len(str2):
            ans = __check_one_away_equal(str1, str2, max_mistake)
        elif len(str2) > len(str1):
            ans = __check_one_away_small(str1, str2, max_mistake)
        else:
            ans = __check_one_away_small(str2, str1, max_mistake)

    return ans


assert check_one_away('pale', 'ple')
assert check_one_away('pale', 'ple')
assert check_one_away('pale', 'ple')
assert check_one_away('pale', 'bale')
assert check_one_away('pale', 'bake') == False
```

```python
### Solution for Question 1.6 ###

def compress(text):
    length = len(text)
    if length == 0:
        return text

    prev_char = text[0]
    repetition = 1
    new_text = ""
    for i in range(1, length):
        c = text[i]
        if c == prev_char:
            repetition += 1
        else:
            new_text = "{}{}{}".format(new_text, prev_char, repetition)
            repetition = 1
            prev_char = c

    # last char
    new_text = "{}{}{}".format(new_text, prev_char, repetition)

    return new_text if len(new_text) < length else text

assert compress('aaaaaaa') == 'a7'
assert compress('a') == 'a'
assert compress('') == ''
assert compress('aaaabbbb') == 'a4b4'
assert compress('aabbc') == 'aabbc'
assert compress('aaabbbc') == 'a3b3c1'
```

```python
### Solution for Question 1.8 ###

def init_matrix(m, n):
    import numpy as np
    return np.random.randint(0, 10, size=(m, n)).tolist()

def zero_seen_at(matrix, i, m, n):
    zeros_at = []
    for j in range(i, n):
        if matrix[i][j] == 0:
            zeros_at.append((i, j))

    for j in range(i, m):
        if matrix[j][i] == 0:
            zeros_at.append((j, i))

    return zeros_at

def fill_with_zero(matrix, i, j, m, n):
    for k in range(m):
        matrix[k][j] = 0

    for k in range(n):
        matrix[i][k] = 0


def fill_zero_if_any(matrix):
    m, n = len(matrix), len(matrix[0])
    zero_coord = []
    for i in range(min(m, n)):
        zero_coord.extend(zero_seen_at(matrix, i, m, n))
            
    for i, j in zero_coord:
        fill_with_zero(matrix, i, j, m, n)
 
def print_func(matrix):
    for arr in matrix:
        print(arr)
    
m = init_matrix(7, 7)
print_func(m)
print('----')
fill_zero_if_any(m)
print_func(m)
```
