```python
import random


def print_nodes(head):
    n = head
    while n is not None:
        print("{} -> ".format(n.data), end='')
        n = n.next
    print("None")

def create_singly_linked_list(n, start=1, end=3):
    nodes = [Node(random.randint(start, end)) for _ in range(n)]
    for i in range(1, n):
        nodes[i-1].next = nodes[i]
    return nodes[0]

def create_singly_linked_list_from_list(elements):
    n = len(elements)
    nodes = [Node(data=e) for e in elements]

    for i in range(1, n):
        nodes[i-1].next = nodes[i]
    return nodes[0]

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
        
def apply_linked_list(head, func):
    n = head
    result = []
    while n is not None:
        result.append(func(n.data))
        n = n.next
    return result
```

```python
head = create_singly_linked_list(9)
print_nodes(head)

# traverse
def remove_dups(head):
    n = head
    prev = head
    nodes_processed = set()
    while n is not None:
        if n.data in nodes_processed:
            prev.next = n.next
        else:
            nodes_processed.add(n.data)
            prev = n
        n = n.next
  
remove_dups(head)
print_nodes(head)
```

```python
# Solution for problem 2.5

def sum_linked_lists(head1, head2):
    n1 = head1
    n2 = head2
    result = ""
    remain = 0 
    while True:
        if n1 is None and n2 is None:
            if remain != 0:
                result = "1{}".format(result)
            return int(result)
        else:
            if n1 is None:
                number1 = 0
            else:
                number1 = n1.data
                n1 = n1.next

            if n2 is None:
                number2 = 0
            else:
                number2 = n2.data
                n2 = n2.next

            total = number1 + number2 + remain
            remain = total // 10
            result = '{}{}'.format(total % 10, result)

# Solution for problem 2.5 - Follow - up

def sum_linked_lists_recur_aux(head1, head2):
    if head1 is None:
        if head2 is None:
            return 0, ""
        else:
            res, t = sum_linked_lists_recur_aux(head1, head2.next)
            return 0, "{}{}".format(head2.data, t)
            
    else:
        if head2 is None:
            res, t = sum_linked_lists_recur_aux(head1.next, head2)
            return 0, "{}{}".format(head1.data, t)
        else:
            res, t = sum_linked_lists_recur_aux(head1.next, head2.next)
            total = head1.data + head2.data + res
            return total // 10, "{}{}".format(total % 10, t)

        
def sum_linked_lists_recur(head1, head2):
    res, t = sum_linked_lists_recur_aux(head1, head2)
    if res != 0:
        t = "1{}".format(t)
    return int(t)
        
        
def get_size(head):
    n = head
    size = 0
    while n is not None:
        size += 1
        n = n.next
    return size


def add_padding(head, num, padding=0):
    n = head
    for i in range(num):
        new_head = Node(0)
        new_head.next = n
        n = new_head
    
    return n
      
    
def equalize_size(head1, head2):
    size1 = get_size(head1)
    size2 = get_size(head2)
    
    if size1 < size2:
        head1 = add_padding(head1, size2-size1, 0)
    elif size2 < size1:
        head2 = add_padding(head2, size1-size2, 0)

    return head1, head2
    
        
head1 = create_singly_linked_list(5, 0, 9)
print_nodes(head1)
head2 = create_singly_linked_list(2, 0, 9)
print_nodes(head2)
print(sum_linked_lists(head1, head2))
print('\nFollow-up Question for same linked lists')
head1, head2 = equalize_size(head1, head2)
print_nodes(head1)
print_nodes(head2)
print(sum_linked_lists_recur(head1, head2))
```

```python
# Solution for problem 2.6

# stack based solution

class Stack(list):
    
    def push(self, value):
        self.append(value)

def is_palindrome_with_stack(head):
    
    n = head
    stack = Stack()
    while n is not None:
        stack.push(n.data)
        n = n.next
    
    n = head
    while n is not None:
        if n.data != stack.pop():
            return False
        n = n.next    
    return True



# Shorter but a bit complicated implementation. 
def is_palindrome_with_stack_short(head):
    stack = apply_linked_list(head, lambda x: x)
    def is_matched(x):
        return x == stack.pop()
    matching = apply_linked_list(head, is_matched)
    return all(matching)



# implementation with recursion
def is_palindrome_recursive(curr_head, orig_head):
    if curr_head.next is None:
        res = curr_head.data == orig_head.data
    else:
        res, orig_head = is_palindrome_recursive(curr_head.next, orig_head)
        res = res and (orig_head.data == curr_head.data)
        
    return res, orig_head.next
       
    
# Some tests

head = create_singly_linked_list_from_list(['1', '2', '3', '2', '1'])
print_nodes(head)

assert is_palindrome_with_stack(head)
assert is_palindrome_recursive(head, head)[0]
assert is_palindrome_with_stack_short(head)

head = create_singly_linked_list_from_list(['1', '2', '3', '2', '3'])
print_nodes(head)

assert is_palindrome_with_stack(head) == False
assert is_palindrome_recursive(head, head)[0] == False
assert is_palindrome_with_stack_short(head) == False

head = create_singly_linked_list_from_list(['1', '5' ,'5', '5', '1'])
print_nodes(head)  

assert is_palindrome_with_stack(head)
assert is_palindrome_recursive(head, head)[0]
assert is_palindrome_with_stack_short(head)
```

### Solutions for Problem 2.7

I have couple of solutions: 
1. color linked lists. One linked list color each node by yellow, other by red (Galatasaray). If we first time hit yellow from red, or red from yellow that's the intersection.
2. two stack solutions (Implemented)
3. When we skip some additional elements in longer linked list, we can iterate both linked list at the same time and compare their values. (implemented)
4. Recursive solution. After checking the number of elements and go through additional elements for longer linked list, we can start do recursive solution. (implemented)

```python
# Solutions for Problem 2.7

# 2. Stack solution.
def intersection_two_stack(head1, head2):
    
    stack1 = Stack()
    n = head1
    while n is not None:
        stack1.push(n)
        n = n.next

    stack2 = Stack()
    n = head2
    while n is not None:
        stack2.push(n)
        n = n.next

    n = None
    while len(stack1) > 0 and len(stack2) > 0:
        e1 = stack1.pop()
        e2 = stack2.pop()
        if e1 == e2:
            n = e1
        else:
            break

    return (n is not None, n)

def skip_nodes(head, number_of_skip):
    n = head
    for _ in range(number_of_skip):
        n = n.next

    return n

def equalize_sizes(head1, head2):
    size1 = get_size(head1)
    size2 = get_size(head2)
        
    # skip nodes for longer linked list.
    
    if size1 > size2:
        head1 = skip_nodes(head1, size1-size2)
    elif size2 > size1:
        head2 = skip_nodes(head2, size2-size1)
    
    return head1, head2

# 3. skipping with iteration solution
def intersection_with_skip(head1, head2):
    head1, head2 = equalize_size(head1, head2)
    while head1 is not None or head2 is not None:
        if head1 == head2:
            return True, head1
        head1 = head1.next
        head2 = head2.next
    
    return False, None


# 4. Recursive Solution
def intersection_with_recursive_aux(head1, head2):
    if head1.next is None or head2.next is None:
        if head1 == head2:
            return head1
        else:
            return None
    else:
        node = intersection_with_recursive_aux(head1.next, head2.next)
        if head1 != head2:
            return node
        else:
            return head1
            
    
        
def intersection_with_recursive(head1, head2):
    # find out sizes (lengths)
    size1 = get_size(head1)
    size2 = get_size(head2)
        
    # skip nodes for longer linked list.
    
    if size1 > size2:
        head1 = skip_nodes(head1, size1-size2)
    elif size2 > size1:
        head2 = skip_nodes(head2, size2-size1)
        
    node = intersection_with_recursive_aux(head1, head2)
    return node is not None, node

# Tests

head1 = create_singly_linked_list(14, 1, 200)
head2 = create_singly_linked_list(14, 1, 200)
head2.next.next.next.next.next.next.next = head1.next.next.next.next

print_nodes(head1)
print_nodes(head2)

answer, node = intersection_two_stack(head1, head2)
print(answer, "|| Should be True", node.data, '')
answer, node = intersection_with_skip(head1, head2)
print(answer, "|| Should be True", node.data, '')
answer, node = intersection_with_recursive(head1, head2)
print(answer, "|| Should be True", node.data, '\n')

head1 = create_singly_linked_list(14, 1, 200)
head2 = create_singly_linked_list(14, 1, 200)

print_nodes(head1)
print_nodes(head2)

answer, node = intersection_two_stack(head1, head2)
print(answer, "|| Should be False")
answer, node = intersection_with_skip(head1, head2)
print(answer, "|| Should be False")
answer, node = intersection_with_recursive(head1, head2)
print(answer, "|| Should be False\n")


head1 = create_singly_linked_list(6, 1, 200)
node = Node('X')
node.next = head1
node1 = Node('XX')
node1.next = node 
head2 = node1

print_nodes(head1)
print_nodes(head2)

answer, node = intersection_two_stack(head1, head2)
print(answer, "|| Should be True", node.data)
answer, node = intersection_with_skip(head1, head2)
print(answer, "|| Should be True", node.data, '')
answer, node = intersection_with_recursive(head1, head2)
print(answer, "|| Should be True", node.data, '\n')
```
